```python
# Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Titanic dataset from a local file using Pandas
titanic = pd.read_excel('/content/titanic3.xls')  # Adjust the file path as needed

# Preview the first few rows of the dataset
print(titanic.head())

# Setting a style for the plots
sns.set_style("whitegrid")

# 1. Bar plot: Survival Count by Gender
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', hue='survived', data=titanic, palette='Set2')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

# 2. Histogram: Age Distribution of Passengers
plt.figure(figsize=(8, 6))
sns.histplot(titanic['age'].dropna(), kde=True, bins=30, color='blue')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 3. Box Plot: Fare Distribution by Passenger Class
plt.figure(figsize=(8, 6))
sns.boxplot(x='pclass', y='fare', data=titanic, palette='Set1')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()

# 4. Scatter Plot: Age vs Fare Colored by Survival Status
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic, palette='Set1')
plt.title('Age vs Fare Colored by Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

# 5. Heatmap: Correlation Matrix of Titanic Dataset (Numeric Columns Only)
plt.figure(figsize=(10, 6))

# Selecting only numeric columns for the correlation matrix
numeric_columns = titanic.select_dtypes(include=['float64', 'int64'])

# Handling missing values by filling them with the mean (you could also drop NaN values)
numeric_columns = numeric_columns.fillna(numeric_columns.mean())

# Computing the correlation matrix
corr_matrix = numeric_columns.corr()

# Plotting the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Titanic Features')
plt.show()

# 6. Violin Plot: Age Distribution by Survival and Gender
plt.figure(figsize=(8, 6))
sns.violinplot(x='sex', y='age', hue='survived', data=titanic, split=True, palette='Set2')
plt.title('Age Distribution by Survival and Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```
